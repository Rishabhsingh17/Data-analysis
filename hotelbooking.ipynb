```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('hotel_booking.csv')
```

```python
columns_to_drop = ['name','email','phone-number','credit_card']
df = df.drop(columns=columns_to_drop,errors='ignore') #deleting unwanted columns
```

```python
df.info()
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date']) # changing datatype 
```

```python
df.info()
```

```python
df.describe(include='object') # it gives total count of non-null values, unique values in a column, most frequent values and frequency of most frequent value.
```

```python
for col in df.describe(include='object').columns:
    print(col)
    print(df[col].unique()) # it gives unique value of all columns
```

```python
df.isnull().sum() # checking for total null values 
```

```python
df.drop(['company','agent'],axis=1,inplace=True,errors='ignore')
df.dropna(inplace=True) #removing all null values
```

```python
df.isnull().sum()
```

```python
df.describe() # getting some mathemetical values.
```

```python
df=df[df['adr']<5000] # creates a database where the value of adr column is less then 5000
```

```python
canceled_percent = df['is_canceled'].value_counts(normalize = True)*100
canceled_percent # getting cancelled percent
```

```python
plt.bar(['not_canceled','canceled'],df['is_canceled'].value_counts())
plt.title("Reservation Status Count") #graph for cancel and non canceled with reservation status
```

```python
ax1=sns.countplot(x='hotel',hue='is_canceled',data = df)
plt.legend(['not_canceled','canceled']) #graph showing cancel and non canceled hotel based on hotel type
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].replace({0:'Not Cancelled',1:'Canceled'}).value_counts(normalize = True)*100 #resort hotel cancellation
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].replace({0:'Not Cancelled',1:'Canceled'}).value_counts(normalize = True)*100 #city hotel cancellation percent
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(30,10))
plt.title('Average daily rate in city and resort hotel',fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
plt.legend(['Not canceled ','canceled'])
plt.title('Reservation status per month')
plt.figure(figsize=(20,10))
plt.show() # finding per month cancellatiion
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.title('ADR Per Month',fontsize=20)
plt.show()
```

```python
canceled_data=df[df['is_canceled']==1]
top_10_countries=canceled_data['country'].value_counts()[:10]#gives value in descending order
plt.figure(figsize=(10,10))
plt.pie(top_10_countries,autopct='%.2f',labels=top_10_countries.index)
plt.show() # cancellation percent per country
```

```python
df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize=True)*100
```

```python
canceled_data['market_segment'].value_counts(normalize=True)*100
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data=df[df['is_canceled']==0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,10))
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label='not cancelled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='cancelled')
plt.legend()
plt.show() # graph showning cancellation yearly
```

```python
canceled_df_adr=canceled_df_adr[(canceled_df_adr['reservation_status_date']>'2016') & (canceled_df_adr['reservation_status_date']<'2017-09')]
not_canceled_df_adr=not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date']>'2016') & (not_canceled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,10))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')
plt.legend(fontsize=20)
```
